```python
if df['Aggregate rating'].dtype != 'float64':
    df['Aggregate rating'] = df['Aggregate rating'].astype(float)
```

```python
df_filtered = df[df['Aggregate rating'] > 0]
```

```python
import seaborn as sns
# Analyze the distribution of the target variable "Aggregate rating"
plt.figure(figsize=(12, 6))
sns.histplot(df_filtered['Aggregate rating'], bins=20, kde=True)
plt.title('Distribution of Aggregate Rating (Filtered)')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
rating_counts_filtered = df_filtered['Aggregate rating'].value_counts()
plt.figure(figsize=(14, 7))
sns.countplot(x='Aggregate rating', data=df_filtered, order=rating_counts_filtered.index)
plt.title('Class Distribution of Aggregate Rating (Filtered)')
plt.xlabel('Aggregate Rating')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate basic statistical measures for numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
statistics = df[numerical_columns].describe().T

# Add median to the statistics dataframe
statistics['median'] = df[numerical_columns].median()
# Calculate standard deviation
std_dev = df['Country Code'].std()
print(f"Standard Deviation: {std_dev}")

# Display the statistics
print(statistics)
```

```python
# Explore the distribution of the "Country Code" variable
country_counts = df['Country Code'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=country_counts.index, y=country_counts.values)
plt.title('Distribution of Country Codes')
plt.xlabel('Country Code')
plt.ylabel('Number of Restaurants')
plt.show()

# Explore the distribution of the "City" variable
city_counts = df['City'].value_counts().head(20)  # Show top 20 cities
plt.figure(figsize=(12, 6))
sns.barplot(x=city_counts.index, y=city_counts.values)
plt.title('Top 20 Cities with the Highest Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=90)
plt.show()

# Explore the distribution of the "Cuisines" variable
cuisines_counts = df['Cuisines'].value_counts().head(20)  # Show top 20 cuisines
plt.figure(figsize=(12, 6))
sns.barplot(x=cuisines_counts.index, y=cuisines_counts.values)
plt.title('Top 20 Cuisines')
plt.xlabel('Cuisines')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=90)
plt.show()

# Identify the top cuisines with the highest number of restaurants
top_cuisines = cuisines_counts.head(10)
print("Top 10 Cuisines with the Highest Number of Restaurants:")
print(top_cuisines)

# Identify the top cities with the highest number of restaurants
top_cities = city_counts.head(10)
print("Top 10 Cities with the Highest Number of Restaurants:")
print(top_cities)
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap


# Filter the data to remove rows with missing latitude or longitude
data = df.dropna(subset=['Latitude', 'Longitude'])

# Create a map centered around the average coordinates
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=2)

# Add points to the map
for _, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"{row['Restaurant Name']} - Rating: {row['Aggregate rating']}"
    ).add_to(restaurant_map)

# Save the map to an HTML file
restaurant_map.save('restaurant_map.html')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of restaurants across countries
country_counts = df['Country Code'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=country_counts.index, y=country_counts.values)
plt.title('Distribution of Restaurants Across Countries')
plt.xlabel('Country Code')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Distribution of restaurants across cities
city_counts = data['City'].value_counts().head(20)  # Show top 20 cities
plt.figure(figsize=(12, 6))
sns.barplot(x=city_counts.index, y=city_counts.values)
plt.title('Top 20 Cities with the Highest Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot of latitude and longitude colored by aggregate rating
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Longitude', y='Latitude', hue='Aggregate rating', palette='viridis', data=data)
plt.title('Restaurant Locations Colored by Aggregate Rating')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

# Calculate correlation matrix
correlation_matrix = data[['Latitude', 'Longitude', 'Aggregate rating']].corr()
print("Correlation Matrix:")
print(correlation_matrix)
```
